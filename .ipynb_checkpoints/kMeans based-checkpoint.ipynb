```python
flat_image = image.reshape((-1, 3))
```

```python
from sklearn.cluster import KMeans
```

```python
n_clusters = 8
comp = KMeans(n_clusters=n_clusters)
```

```python
cluster_idx = comp.fit_predict(flat_image)
```

```python
clusters = comp.cluster_centers_
```

```python
flat_image_compressed = np.array([clusters[ci] for ci in cluster_idx])
```

```python
image_reconstructed = np.clip(flat_image_compressed.reshape((image.shape[0], image.shape[1], 3)), 0, 255).astype(np.int32)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.title('Reconstructed')
plt.imshow(image_reconstructed)
```

```python
def kmeans_comp(image, n_clusters):
    comp = KMeans(n_clusters=n_clusters)
    cluster_idx = comp.fit_predict(flat_image)
    clusters = comp.cluster_centers_
    flat_image_compressed = np.array([clusters[ci] for ci in cluster_idx])
    image_reconstructed = np.clip(flat_image_compressed.reshape((image.shape[0], image.shape[1], 3)), 0, 255).astype(np.int32)

    return image_reconstructed
```

```python
def kmeans_compress(image, n_clusters):
    flat_image = image.reshape((-1, 3))
    comp = KMeans(n_clusters=n_clusters)
    cluster_idx = comp.fit_predict(flat_image)
    clusters = comp.cluster_centers_
    flat_image_compressed = np.array([clusters[ci] for ci in cluster_idx])
    image_reconstructed = np.clip(flat_image_compressed.reshape((image.shape[0], image.shape[1], 3)), 0, 255).astype(np.uint8)

    return image_reconstructed
```

```python
nC = [10, 20, 50]

plt.figure(figsize=(16, 16))

plt.subplot(2, 2, 1)
plt.title('Original')
plt.imshow(image)

for i, n_clusters in enumerate(nC):
    plt.subplot(2, 2, i+2)
    plt.title(n_clusters)
    
    image_reconstructed = kmeans_comp(image, n_clusters)
    
    plt.imshow(image_reconstructed)
```

```python
path = 'results/kmean_{}_{}.jpeg'
nC = [3]

for i in idxs:
    for n_clusters in nC:
        image_reconstructed = kmeans_compress(images[i], n_clusters)
        plt.imsave(path.format(i, n_clusters), image_reconstructed)
```
