```python
f_tiles = [scipy.fft.fft2(x) for x in tiles]

fft_hat_tiles = []

for f_t in f_tiles:
    fft_hat = [None, None, None]
    for c in range(3):
        
        ener = np.abs(f_t[..., c]).reshape((-1, ))
        total_ener = np.sum(ener)

        thres = 0.95

        ener_idx = np.argsort(ener)[::-1]

        cumulative_var = np.cumsum(ener[ener_idx])

        i = np.where(cumulative_var > total_ener*thres)[0].min()

        fft_hat[c] = f_t[..., c]

        for j in range(i, len(ener_idx)):
            fft_hat[c][ener_idx[j]//fft_hat[c].shape[0], ener_idx[j]%fft_hat[c].shape[1]] = 0+0*1j
    fft_hat = np.array(fft_hat).astype(np.csingle)
    fft_hat_tiles.append(np.moveaxis(fft_hat, [0, 1, 2], [2, 0, 1]))
```

```python


def transform_coding(im, M=8, N=8, T=scipy.fft.fft2, T_inv=scipy.fft.ifft2, thres = 0.95):
    tiles = [im[x:x+M,y:y+N] for x in range(0,im.shape[0],M) for y in range(0,im.shape[1],N)]

    f_tiles = [T(x) for x in tiles]

    fft_hat_tiles = []

    for f_t in f_tiles:
        fft_hat = [None, None, None]
        for c in range(3):

            ener = np.abs(f_t[..., c]).reshape((-1, ))
            total_ener = np.sum(ener)

            

            ener_idx = np.argsort(ener)[::-1]

            cumulative_var = np.cumsum(ener[ener_idx])

            i = np.where(cumulative_var > total_ener*thres)[0].min()

            fft_hat[c] = f_t[..., c]

            for j in range(i, len(ener_idx)):
#                 print(fft_hat[c])
                fft_hat[c][ener_idx[j]//fft_hat[c].shape[0], ener_idx[j]%fft_hat[c].shape[1]] = 0
        fft_hat = np.array(fft_hat).astype(np.csingle)
        fft_hat_tiles.append(np.moveaxis(fft_hat, [0, 1, 2], [2, 0, 1]))

    tiles_hat = [np.clip(np.real(T_inv(x)), 0, 255).astype(np.int8) for x in fft_hat_tiles]

    image_hat = np.zeros_like(im)

    k = 0
    for x in range(0,im.shape[0],M):
        for y in range(0,im.shape[1],N):
            image_hat[x:x+M,y:y+N] = tiles_hat[k]
            k += 1
    return image_hat
```

```python
idxs = [11170, 10609, 17689, 14521, 3994, 14633, 1938, 12871, 9254, 10513, 9871]
```

```python
path = 'results/TC_{}_{}.jpeg'
t = ['fft', 'dct', 'dst', 'fht']
t_func = [scipy.fft.fft2, lambda x: scipy.fft.dct(scipy.fft.dct(x.T).T),
          lambda x: scipy.fft.dst(scipy.fft.dst(x.T).T),
          lambda x: scipy.fft.fht(scipy.fft.fht(x.T).T)
         ]

t_inv_func = [scipy.fft.ifft2, lambda x: scipy.fft.idct(scipy.fft.idct(x.T).T),
          lambda x: scipy.fft.idst(scipy.fft.idst(x.T).T),
          lambda x: scipy.fft.ifht(scipy.fft.ifht(x.T).T)
         ]

thres = 0.98
M, N = 4, 4

for i in idxs:
    for (ts, T, T_inv) in zip(t, t_func, t_inv_func):
        if ts == 'fht':
            continue
        image_reconstructed = transform_coding(images[i], M=M, N=N, T=T, T_inv=T_inv, thres=thres)
        plt.imsave(path.format(i, ts), image_reconstructed)
```
